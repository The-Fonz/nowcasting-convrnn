```python
import numpy as np
import torch
from torch.autograd import Variable
import torch.nn as nn

from config_vpn import model

model.decoder.mask()

def b(conv):
    "Binarize weights"
    w = conv.weight.data.cpu().numpy()
    w = (w!=0).astype(np.uint8)
    return w

def print_tensor_2d(t, indx=tuple()):
    "Recursive print function that prints per row"
    if t.dim() == 2:
        print(indx)
        for row in t:
            print(' '.join('{:.2f}'.format(d) for d in row))
    else:
        for i, r in enumerate(t):
            print_tensor_2d(r, indx=indx + (i,))

def print_tensor_list(l, pix=None):
    for i, t in enumerate(l):
        if pix:
            for i_ts in range(t.data.size(0)):
                print('({:d}, {:d})'.format(i, i_ts))
                print('\n'.join('{:.2f}'.format(d) for d in t.data[i_ts, :, pix[0], pix[1]]))
        else:
            print(i)
            print_tensor_2d(t.data)
```

```python
def weight_init(l):
    if isinstance(l, nn.Conv2d):
#         print(l.weight.data)
        # In-place
        l.weight.data.fill_(1)
        if l.bias is not None:
            l.bias.data.fill_(0)
#         print(l.weight.data)

model.apply(weight_init);
model.decoder.mask()
```

```python
history = Variable(torch.zeros(1,1,46,12,12), volatile=True)
x = Variable(torch.zeros(1,1,1,12,12), volatile=True)

x[0,0,0,1,1] = 1

model.train()
print_tensor_list(model.decoder(history, targets=x, return_intermediate=True), pix=(1,1))
```
